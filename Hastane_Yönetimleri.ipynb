```python
import random
import time

print("Merhaba, hastane hizmetlerine erişmek için lütfen size verilen şifreyi giriniz...")
sifre_girisi = input("Şifre : ")

#DCÜHYSG -- Doğan Cüceloğlu Üniversite Hastanesi Yönetim Sistemi Girişi

if sifre_girisi == "DCÜHYSG":
  time.sleep(1)
  print("Merhaba, @Yönetim, Yapmak İstediğiniz işlemi seçin...")
  i_s = input("HK-Hasta Kaydı , HTE-Hasta Taburcu Etme , HONG-Hasta Oda Numarası Görüntüleme, CIKIS-Sistemden Çıkış")

#Büyük harflerle 'HK','HTE','HONG' veya 'CIKIS' yazılmalı yoksa hata verir.(Otomatik geçersiz girişe yönlendirir...)

  if i_s == "HK":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(i_s))
    time.sleep(1)
    h_k_g = input("Kaydını yapmak istediğiniz hastanın ismi: ")
    h_k_g_s = input("Kaydını yapmak istediğiniz hastanın soyismi: ")
    h_v_o_n = random.randrange(1,5000)
    time.sleep(1)
    print("{} isimli, {} soyisimli hastanın sisteme girişi sağlandı. \nHastaya verilen oda numarası: {}".format(h_k_g, h_k_g_s, h_v_o_n))

  elif i_s == "HTE":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(i_s))
    time.sleep(1)
    h_t_g = input("Taburcu etmek istediğiniz hastanın ismi: ")
    h_t_g_s = input("Taburcu etmek istediğiniz hastanın soyismi: ")
    h_v_o_n_t = random.randrange(1,5000)
    time.sleep(1)
    print("{} isimli, {} soyisimli hastanın sistemden çıkışı sağlandı. \nÇıkışı yapılan hastanın oda numarası: {}".format(h_t_g, h_t_g_s, h_v_o_n_t))

  elif i_s == "HONG":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(i_s))
    time.sleep(1)
    h_o_g = input("Oda numarasını görüntülemek istediğiniz hastanın ismi: ")
    h_o_g_s = input("Oda numarasını görüntülemek istediğiniz hastanın soyismi: ")
    h_v_o_n_o = random.randrange(1,5000)
    time.sleep(1)
    print("{} isimli, {} soyisimli hastanın oda numarası bulundu. \nHastanın oda numarası: {}".format(h_o_g, h_o_g_s, h_v_o_n_o))

  elif i_s == "CIKIS":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(i_s))
    print("Başarıyla çıkış yapıldı...")
  
  else:
    time.sleep(1)
    print("Üzgünüz @Yönetim fakat girdiğiniz işlem verilenler ile uyumlu değil...")

#DCÜHHHG -- Doğan Cüceloğlu Üniversite Hastanesi Hasta Hakları Girişi

elif sifre_girisi == "DCÜHHHG":
  time.sleep(1)
  print("Merhaba @Hasta. Lütfen işlemlerden birini seçiniz.")
  h_i_s = input("HÇ-Hemşire/Hemşir Çağırma, DÇ-Doktor Çağırma, KIH-Kendimi İyi Hissetmiyorum, CIKIS-Sistemden Çıkış")

  if h_i_s == "HÇ":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(h_i_s))
    time.sleep(1)
    print("Odanıza en yakın Hemşir/Hemşire çağırıldı. \nSağlıklı günler dileriz!")

  elif h_i_s == "DÇ":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(h_i_s))
    time.sleep(1)
    print("Odanıza en yakın Doktor çağırıldı. \nSağlıklı günler dileriz!")

  elif h_i_s == "KIH":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(h_i_s))
    time.sleep(1)
    print("Durumunuz, hastalığınızın önem derecesine göre en yakın doktora ve hemşir ve hemşirelere haber verildi. \nSağlıklı günler dileriz!")

  elif h_i_s == "CIKIS":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(h_i_s))
    time.sleep(1)
    print("Sistemden başarı ile çıkıldı. \nSağlıklı günler dileriz!")

  else:
    time.sleep(1)
    print("Seçtiğiniz işlem belirtilenler ile uyuşmuyor.")

#DCÜHPHG -- Doğan Cüceloğlu Üniversite Hastanesi Personel Hakları Girişi

elif sifre_girisi == "DCÜHPHG":
  time.sleep(1)
  print("Merhaba @Personel, lütfen işlem seçiniz!")
  p_i_g = input("MNZY-Maaşlar Ne Zaman Yatıyor, \nMTNK-Maaş Tutarı Ne Kadar, \nCIKIS-Sistemden Çıkış")

  if p_i_g == "MNZY":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(p_i_g))
    time.sleep(1)
    print("Maaşlar her ayın ilk 5 günü içerisinde düzenli olarak verilir. \nİyi günler dileriz!")

  elif p_i_g == "MTNK":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(p_i_g))
    time.sleep(1)
    print("Maaşlar yeni bir güncelleme gelene kadar '3850 TL'dir. \nİyi günler dileriz!")

  elif p_i_g == "CIKIS":
    time.sleep(1)
    print("Seçtiğiniz işleme giriş sağlandı, seçtiğiniz işlem: {} ".format(p_i_g))
    time.sleep(1)
    print("Sistemden başarı ile çıkıldı. \nİyi günler dileriz!")

  else:
    time.sleep(1)
    print("Üzgünüz ancak, girdiğiniz işlem belirtilenler ile uyuşmuyor.")

else:
  time.sleep(1)
  print("Üzgünüz ancak girdiğiniz şifre hiçbir giriş ile uyumlu değil... \nSağlıklı günler dileriz!")
```
