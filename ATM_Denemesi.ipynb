```python
import time
#time'ı yükleniyor hissi vermek için kullanıyorum :)

import random
#randomu başlangıçtaki para miktarı için kullanıyorum :)

kalanpara = random.randrange(1,50000)

print("Merhaba @Ege. Banka hesabınıza erişim sağlamak için lütfen şifrenizi giriniz.")
user_sifre = int(input("Şifreniz: "))

if user_sifre == 9136:
  print("Hesabınıza başarı ile giriş sağlandı. Hesabınızda kalan para: {}".format(kalanpara))
  time.sleep(1)
  print("Hesabınıza şu anki girişiniz ile 5 kez işlem yapabileceksiniz.")
  
  for i in range (5):
    kullanıcı_islem = input("İşlemlerden birini seçiniz: \nPÇ-Para Çekme, \nPY-Para Yatırma : ")
    print("Yükleniyor...")
    time.sleep(1)
    

    if kullanıcı_islem == "PÇ":
      print("Yükleniyor...")
      time.sleep(1)
      print("Seçtiğiniz işleme başarı ile giriş sağlandı. Seçtiğiniz işlem: 'Para Çekme'")
      PÇ = int(input("Çekmek istediğiniz para miktarını lütfen giriniz: "))
      kalanpara = kalanpara - PÇ
      print("Yükleniyor...")
      time.sleep(1)
      print("Hesabınızdan şu miktarda para çekildi: {}, Hesabınızda kalan para miktarı: {}".format(PÇ, kalanpara))
      borc_sorgu_1 = input("Borcunuzu sorgulamak ister misiniz? E/H : ")
      borc_miktarı_1 = kalanpara*(-1)

      if borc_sorgu_1 == "E":
        print("Yükleniyor...")
        time.sleep(1)

        if kalanpara < 0:
          print("Hesabınız borçludur veya borçlu olmuştur. Borçlu olduğunuz para miktarı : ", borc_miktarı_1, "₺")
          print("Yükleniyor...")
          time.sleep(1)

        else:
          print("Hesabınız borçlu değildir veya borcu kapanmıştır.")
          print("Yükleniyor...")
          time.sleep(1)

      elif borc_sorgu_1 == "H":
        print("Yükleniyor...")
        time.sleep(1)

    elif kullanıcı_islem == "PY":
      print("Yükleniyor...")
      time.sleep(1)
      print("Seçtiğiniz işleme başarı ile giriş sağlandı. Seçtiğiniz işlem: 'Para Yatırma'")
      PY = int(input("Yatırmak istediğiniz para miktarını lütfen giriniz: "))
      kalanpara = kalanpara + PY
      print("Yükleniyor...")
      time.sleep(1)
      print("Hesabınıza şu miktarda para yatırıldı: {}, Hesabınızda kalan para miktarı: {}".format(PY, kalanpara))
      borc_sorgu_2 = input("Borcunuzu sorgulamak ister misiniz? E/H : ")
      borc_miktarı_2 = kalanpara*(-1)

      if borc_sorgu_2 == "E":
        print("Yükleniyor...")
        time.sleep(1)

        if kalanpara < 0:
          print("Hesabınız borçludur veya borçlu olmuştur. Borçlu olduğunuz para miktarı : ", borc_miktarı_2, "₺")
          print("Yükleniyor...")
          time.sleep(1)

        else:
          print("Hesabınız borçlu değildir veya borcu kapanmıştır.")
          print("Yükleniyor...")
          time.sleep(1)

      elif borc_sorgu_2 == "H":
        print("Yükleniyor...")
        time.sleep(1)

      else:
        print("Yükleniyor...")
        time.sleep(1)
        print("Lütfen 'E' ve 'H' değerlerinden birini girdiğinizden veya büyük harf ile girdiğinizden emin olun ve sonra tekrar deneyin.")

    else:
      print("Yükleniyor...")
      time.sleep(1)
      print("Seçtiğiniz işlem belirtilenler ile uyuşmadı.")
  
  print("Yükleniyor...")
  time.sleep(1)
  print("Hesabınızdan çıkış sağlandı. Hesabınızdaki para miktarı : ", kalanpara)

  
else:
  print("Yükleniyor...")
  time.sleep(1)
  print("Girdiğiniz şifre veriler ile uyumlu değil.")
```
